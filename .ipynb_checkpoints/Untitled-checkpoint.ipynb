```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do novo arquivo CSV usando a codificação 'latin1' e delimitador ';'
file_path = 'dpnibr17196776287.csv'  # Coloque o nome do arquivo ou o caminho relativo
data = pd.read_csv(file_path, encoding='latin1', delimiter=';')

# Exibir as primeiras linhas dos dados para inspeção
print("Dados brutos:")
print(data.head())

# Remover as primeiras linhas que não contêm dados relevantes
data_cleaned = data.iloc[1:].reset_index(drop=True)

# Renomear colunas
data_cleaned.columns = ['Região'] + list(range(1994, 2023)) + ['Total']

# Selecionar apenas a linha que contém os totais por ano
data_totals = data_cleaned[data_cleaned['Região'] == 'Total']

# Verificar se a coluna 'Total' existe e removê-la junto com 'Região'
columns_to_drop = ['Região']
if 'Total' in data_totals.columns:
    columns_to_drop.append('Total')
data_totals = data_totals.drop(columns=columns_to_drop).T

# Renomear colunas
data_totals.columns = ['Doses Aplicadas']

# Resetar índice e renomear colunas para anos
data_totals.reset_index(inplace=True)
data_totals.columns = ['Ano', 'Doses Aplicadas']

# Converter a coluna 'Ano' para inteiro
data_totals['Ano'] = data_totals['Ano'].astype(int)

# Exibir os dados limpos
print("Dados limpos:")
print(data_totals.head())

# Salvar dados limpos em um novo arquivo CSV
output_path = 'dados_imunizacoes_todas_vacinas_limpos.csv'  # Coloque o nome do arquivo de saída ou o caminho relativo
data_totals.to_csv(output_path, index=False)
```

```python
# Carregar os dados limpos
data_totals = pd.read_csv('dados_imunizacoes_todas_vacinas_limpos.csv')

# Exibir os dados limpos para verificação
print("Dados de Imunizações Limpos - Todas as Vacinas:")
print(data_totals)

# Plotar o gráfico de imunizações ao longo dos anos com escala ajustada para milhões
plt.figure(figsize=(2, 6))
plt.plot(data_totals['Ano'], data_totals['Doses Aplicadas'] / 1e6, marker='o', label='Doses Aplicadas (milhões)')
plt.axvline(x=1994, color='magenta', linestyle='--', label='Erradicação da Poliomielite nas Américas')
plt.axvline(x=1997, color='cyan', linestyle='--', label='Campanha Nacional de Vacinação contra Sarampo')
plt.axvline(x=2000, color='gray', linestyle='--', label='Continuidade de Expansão do PNI')
plt.axvline(x=2002, color='gold', linestyle='--', label='Introdução da Vacina contra Hepatite B')
plt.axvline(x=2003, color='navy', linestyle='--', label='Implementação da Vacina contra Hib')
plt.axvline(x=2004, color='lime', linestyle='--', label='Introdução da Vacina contra Pneumococo')
plt.axvline(x=2006, color='teal', linestyle='--', label='Ampliação da Vacina contra Influenza')
plt.axvline(x=2008, color='cyan', linestyle='--', label='Campanha de Vacinação contra Rubéola')
plt.axvline(x=2010, color='blue', linestyle='--', label='Campanha de Vacinação contra H1N1')
plt.axvline(x=2011, color='purple', linestyle='--', label='Ano Pós-Campanha H1N1')
plt.axvline(x=2013, color='pink', linestyle='--', label='Introdução de Vacinas: HPV, Pentavalente e VIP')
plt.axvline(x=2014, color='red', linestyle='--', label='Início da Crise Econômica')
plt.axvline(x=2016, color='orange', linestyle='--', label='Ponto mais crítico da crise')
plt.axvline(x=2019, color='green', linestyle='--', label='Esforços de Vacinação antes da Pandemia')
plt.axvline(x=2020, color='black', linestyle='--', label='Início da Pandemia de COVID-19')
plt.title('Número de Imunizações ao Longo do Tempo (1994-2022)')
plt.xlabel('Ano')
plt.ylabel('Número de Doses Aplicadas (milhões)')
plt.legend()
plt.grid(True)
plt.show()

# Análise dos Dados
print("Análise dos Dados de Imunização:")
for index, row in data_totals.iterrows():
    ano = row['Ano']
    doses = row['Doses Aplicadas']
    if ano in [1994, 1997, 2000, 2002, 2003, 2004, 2006, 2008, 2010, 2011, 2013, 2014, 2016, 2019, 2020]:
        if ano == 1994:
            evento = "Erradicação da Poliomielite nas Américas"
        elif ano == 1997:
            evento = "Campanha Nacional de Vacinação contra Sarampo"
        elif ano == 2000:
            evento = "Continuidade de Expansão do PNI"
        elif ano == 2002:
            evento = "Introdução da Vacina contra Hepatite B"
        elif ano == 2003:
            evento = "Implementação da Vacina contra Hib"
        elif ano == 2004:
            evento = "Introdução da Vacina contra Pneumococo"
        elif ano == 2006:
            evento = "Ampliação da Vacina contra Influenza"
        elif ano == 2008:
            evento = "Campanha de Vacinação contra Rubéola"
        elif ano == 2010:
            evento = "Campanha de Vacinação contra H1N1"
        elif ano == 2011:
            evento = "Ano Pós-Campanha H1N1"
        elif ano == 2013:
            evento = "Introdução de Vacinas: HPV, Pentavalente e VIP"
        elif ano == 2014:
            evento = "Início da Crise Econômica"
        elif ano == 2016:
            evento = "Ponto mais crítico da crise"
        elif ano == 2019:
            evento = "Esforços de Vacinação antes da Pandemia"
        elif ano == 2020:
            evento = "Início da Pandemia de COVID-19"
        print(f"Ano: {ano}, Doses Aplicadas: {doses}, Evento: {evento}")
    else:
        print(f"Ano: {ano}, Doses Aplicadas: {doses}")
```
